```python
import numpy
import numpy as np
import pandas as pd
import os 
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import cv2
import matplotlib.pyplot as plt
class CycleGANDataset(Dataset):
    def __init__(self):
        self.path='dataset/ICCV15_fashion_dataset(ATR)/humanparsing/JPEGImages/'
        self.images_list = [x for x in os.listdir(self.path) if 'dataset10k' in x][:500]
        # self.labels_list = [x for x in os.listdir('dataset/ICCV15_fashion_dataset(ATR)/humanparsing/SegmentationClassAug') if 'dataset10k' in x][:500]

    def __len__(self):
        return len(self.images_list)
    def __getitem__(self,index):
        img=cv2.imread(self.path+self.images_list[index])
        return img,cv2.stylization(img, sigma_s=60, sigma_r=0.6)
data=CycleGANDataset()
```
